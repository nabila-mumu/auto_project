```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("your_dataset.csv")

# Sort by num_nodes (important for clean lines)
df = df.sort_values(by="num_nodes")

# Plot
plt.figure()

plt.plot(df["num_nodes"], df["bully_time"], marker='o', label="Bully Time")
plt.plot(df["num_nodes"], df["bi-dir_ring_time"], marker='s', label="Bi-directional Ring Time")
plt.plot(df["num_nodes"], df["raft_time"], marker='^', label="Raft Time")

# Labels and title
plt.xlabel("Number of Nodes")
plt.ylabel("Time")
plt.title("Comparison of Election Algorithm Time vs Number of Nodes")

# Legend
plt.legend()

# Grid
plt.grid()

# Show
plt.show()
```
